# matplotlib的使用方法
## 1、导入相应的函数库
matplotlib中常用的pyplot模块，其中包含了许多实用的显示函数，一般简写为plt
安装时，通过conda命令直接安装：

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

### 实例：
通过numpy生成x-y1,x-y2两条曲线，并绘制在同一个坐标轴上

```python
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([1.0, 2.0, 3.0, 5.0, 9.0])
y2 = np.array([0.0, 3.2, 4.8, 7.6, 10.1])
```

通过`plot()`函数可以将x和y值进行对应组合为坐标，绘制在预设版图中，在通过`show()`函数进行显示。

```python
plt.title("x-y1,y2 demo")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

## 2、中文显示
Matplotlib默认情况不支持中文，但可以通过导入字体库的方式来解决：
首先下载字体的.tff文件，放在当前执行的代码文件中（可以指定路径），或者通过fontproperties参数指定系统字体库中的中文字体，例如：`fontproperties="STsong"`

```python
# 导入字体文件，其中fname参数为字体库的路径，必须是.tff字体库
zhfont1 = matplotlib.font_manager.FontProperties(fname="SimHei.ttf")

x = np.arange(1,11)
y = 2 * x * x + 5

# fontproperties 设置中文显示， fontsize 设置字体大小
plt.title("中文字体-测试", fontproperties=zhfont1, fontsize=10)
plt.xlabel("x 轴", fontproperties=zhfont1)
plt.ylabel("y 轴", fontproperties=zhfont1)
plt.plot(x, y)
plt.show()
```

此外，还可以通过使用系统字体的方法。通过以下程序打印出所有注册的tff字体，找到其中一种：STsong（宋体）进行替换

```python
a=sorted([f.name for f in matplotlib.font_manager.fontManager.ttflist])
n=0
for i in a:
    print(i)
    n += 1
    if n == 5:
        break
```

```python
plt.title("中文字体-测试", fontproperties="STsong", fontsize=20)
plt.xlabel("x 轴", fontproperties="STsong")
plt.ylabel("y 轴", fontproperties="STsong")
plt.plot(x, y)
plt.show()
```

## 3、绘制正弦波
通过以下实例使用matplotlib生成正弦波图，只要在字符串前后添加"$"符号，matplotlib就会使用其内嵌的latex引擎绘制的数学公式。

```python
# 计算正弦曲线上点的 横、纵 坐标
sita = np.arange(0,3*np.pi, 0.1)
y = np.sin(sita)
plt.title("sin wave form $sin(x)$")
plt.plot(sita, y, "b")
plt.show()
```

## 4、绘制多个函数图像
`subplot()`函数可以让你在一幅图中绘制多个不同的函数图像，其用法与matlab中的subplot方法相类似。以下实例绘制正弦和余弦图：

```python
# 计算正弦和余弦曲线上的点的 x 和 y 坐标 
x = np.arange(0,  3  * np.pi,  0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
# 建立 subplot 网格，高为 2，宽为 1  
# 激活第一个 subplot
plt.subplot(2,  1,  1)  
# 绘制第一个图像 
plt.plot(x, y_sin) 
plt.title('Sine')  
# 将第二个 subplot 激活，并绘制第二个图像
plt.subplot(2,  1,  2) 
plt.plot(x, y_cos) 
plt.title('Cosine')  
# 展示图像
plt.show()
```

## 5、绘制统计直方图
pyplot子模块提供`bar()`函数来生成条形图，以下实例生成两组x和y数组的条形图

```python
x =  [5,8,10] 
y =  [12,16,6] 
x2 =  [6,9,11] 
y2 =  [6,15,7] 
plt.bar(x, y, align = 'center') 
plt.bar(x2, y2, color = 'g', align = 'center') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis') 
plt.show()
```

`numpy.histogram()`函数是数据的频率分布的图形表示。 水平尺寸相等的矩形对应于类间隔，称为 bin，变量 height 对应于频率。`numpy.histogram()`函数将输入数组和 bin 作为两个参数。 bin 数组中的连续元素用作每个 bin 的边界。

Matplotlib 可以将直方图的数字表示转换为图形。 pyplot 子模块的`plt()`函数将包含数据和 bin 数组的数组作为参数，并转换为直方图。

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
np.histogram(a,bins =  [0,20,40,60,80,100]) 
hist,bins = np.histogram(a,bins = [0,20,40,60,80,100])  
print (hist) 
print (bins)
```

```python
#通过matplotlib中的his（）进行统计直方图（histogram）绘制
plt.hist(a, bins)
plt.title("histogram") 
plt.show()
```

## 6、散点图 
通过`plt.scatter()`函数可以进行散点图的回执，其中常用的参数：s为面积，c为颜色，alpha为透明度，maker为散点标记。

```python
x = np.random.normal(size=500)
y = np.random.normal(size=500)

# s-面积，c-颜色，alpha-透明度，maker-记号
plt.scatter(x,y,s=50,c='b',alpha=0.5)

# 控制坐标轴显示长度
plt.xlim((-2,2))
plt.ylim((-2,2))

#隐藏坐标轴
plt.xticks(())
plt.yticks(())

plt.show()
```

## 7、等高线
在生成等高线中，我们需要三维的数据点：（x，y）和对应的高度。在本程序中，高度我们用一个函数`f(x,y)`来生成。x，y分别是在区间[-3,3]中均匀分布的100个值，并用`plt.meshgrid`在二维平面中将每一个x和每一个y分别对应起来，编织成栅格

```python
def f(x, y):  
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2) 
# x和y在-3,3的范围内平均取100个点
x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)

# 生成网格点坐标矩阵
X,Y = np.meshgrid(x,y)

# 颜色填充
# cmap为颜色模式
plt.contourf(X,Y,f(X,Y),cmap=plt.cm.hot)

# 画等高线
C = plt.contour(X,Y,f(X,Y),colors='black')
plt.clabel(C,fontsize=10)

plt.xticks(())
plt.yticks(())
plt.show()
```
